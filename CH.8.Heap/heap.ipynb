```python
class MaxHeap:
    """"""
    	
    def __init__(self, nums: list[int]):
        """"""
        # 
        self.max_heap = nums
    	
    def left(self, i: int) -> int:
        """"""
        return 2 * i + 1
    
    def right(self, i: int) -> int:
        """"""
        return 2 * i + 2
    	
    def parent(self, i: int) -> int:
        """"""
        return (i - 1) // 2  # 
    
    def size(self) -> int:
        """"""
        return len(self.max_heap)
    	
    def is_empty(self) -> bool:
        """"""
        return self.size() == 0
    	
    def peek(self) -> int:
        """"""
        return self.max_heap[0]
    
    def push(self, val: int):
        """"""
        self.max_heap.append(val)
        self.sift_up(self.size() - 1)
    
    def swap(self, x: int, y: int):
        self.max_heap[x], self.max_heap[y] = self.max_heap[y], self.max_heap[x]
    
    def sift_up(self, i: int):
        #  i 
        while True:
            #  i 
            p = self.parent(i)
            if p < 0 or self.max_heap[i] <= self.max_heap[p]:
                break
            # switch
            self.swap(i, p)
            # 
            i = p     
            
    def pop(self) -> int:
        # 
        if self.is_empty():
            raise IndexError('Null')
        self.swap(0, self.size() - 1)
        # del
        val = self.max_heap.pop()
        self.sift_down(0)
        return val

    def sift_down(self, i: int):
        """i, """
        while True:
            l = self.left(i)
            r = self.right(i)
            if i > self.size() - 1:
                break
            if self.max_heap[i] < self.max_heap[l]:                
                self.swap(i, l)
                i = l
            elif self.max_heap[i] < self.max_heap[r]:                
                self.swap(i, r)
                i = r
            else:
                break    
```
