我们可以将索引映射公式封装成函数，方便后续使用：

```python
def left(self, i: int) -> int:
    return 2 * i + 1

def right(self, i: int) -> int:
    return 2 * i + 2

def parent(self, i: int) -> int:
    return (i - 1) // 2
```

与元素入堆操作相似，堆顶元素出堆操作的时间复杂度也为O(logn)。代码如下所示：

```python
class MaxHeap:
    """大顶堆"""
    	
    def __init__(self, nums: list[int]):
        """构造方法"""
        # 将列表元素原封不动添加进堆
        self.max_heap = nums
    	
    def left(self, i: int) -> int:
        """获取左子节点的索引"""
        return 2 * i + 1
    
    def right(self, i: int) -> int:
        """获取右子节点的索引"""
        return 2 * i + 2
    	
    def parent(self, i: int) -> int:
        """获取父节点的索引"""
        return (i - 1) // 2  # 向下整除
    
    def size(self) -> int:
        """获取堆大小"""
        return len(self.max_heap)
    	
    def is_empty(self) -> bool:
        """判断堆是否为空"""
        return self.size() == 0
    	
    def peek(self) -> int:
        """访问堆顶元素"""
        return self.max_heap[0]
    
    def push(self, val: int):
        """"""
        self.max_heap.append(val)
        self.sift_up(self.size() - 1)
    
    def swap(self, x: int, y: int):
        self.max_heap[x], self.max_heap[y] = self.max_heap[y], self.max_heap[x]
    
    def sift_up(self, i: int):
        # 从节点 i 开始，从底至顶堆化
        while True:
            # 获取节点 i 的父节点
            p = self.parent(i)
            if p < 0 or self.max_heap[i] <= self.max_heap[p]:
                break
            # switch
            self.swap(i, p)
            # 循环向上堆化
            i = p     
            
    def pop(self) -> int:
        # 判空处理
        if self.is_empty():
            raise IndexError('Null')
        self.swap(0, self.size() - 1)
        # del
        val = self.max_heap.pop()
        self.sift_down(0)
        return val

    def sift_down(self, i: int):
        """从节点i开始, 自顶像底堆化"""
        while True:
            l = self.left(i)
            r = self.right(i)
            if i > self.size() - 1:
                break
            if self.max_heap[i] < self.max_heap[l]:                
                self.swap(i, l)
                i = l
            elif self.max_heap[i] < self.max_heap[r]:                
                self.swap(i, r)
                i = r
            else:
                break    
```

```python
"""Driver Code"""
if __name__ == "__main__":
    # 初始化大顶堆
    # 请注意，输入数组已经是一个已经是一个合法的堆
    max_heap = MaxHeap([9, 8, 7, 6, 7, 6, 2, 1, 4, 3, 6, 2, 5])

    # 堆顶元素出堆
    peek = max_heap.pop()
    
    print(max_heap.max_heap)
    print(peek)
```
