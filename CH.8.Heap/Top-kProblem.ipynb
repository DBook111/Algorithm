对于该问题，我们先介绍两种思路比较直接的解法，再介绍效率更高的堆解法。

```python
import heapq

def top_k_heap(nums: list[int], k: int) -> list[int]:
    # 初始化小顶堆
    heap = []
    # 将数组前 k 个元素入堆
    for i in range(k):
        heapq.heappush(heap, nums[i])
    # 从第 k+1 个元素开始入堆，并进行堆化
    for i in range(k, len(nums)):
        if nums[k] > heap[0]:
            heapq.heappop(heap)
            heapq.heappush(heap, nums[i])
    return heap
```

```python
"""Driver Code"""
if __name__ == "__main__":
    nums = [1, 7, 6, 3, 2]
    k = 3

    res = top_k_heap(nums, k)
    print(res)
```
