```python
def bubble_sort(nums: list[int]):
    """bubble"""
    l = len(nums)
    for i in range(l):        
        for j in range(l - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
    return nums
```

```python
"""Driver Code"""
if __name__ == "__main__":
    nums = [4, 1, 3, 1, 5, 2]
    bubble_sort(nums)
    print(" nums =", nums)
```
