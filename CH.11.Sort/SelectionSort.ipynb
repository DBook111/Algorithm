```python
def swap(nums: list[int], i: int, j: int):
    nums[i], nums[j] = nums[j], nums[i]
    return nums

def selection_sort(nums: list[int]):
    l = len(nums)    
    for i in range(l):
        j = i + 1
        min_index = i
        for k in range(j, l):
            if nums[min_index] > nums[k]:
                min_index = k
        nums = swap(nums, min_index, i)        
    return nums
```

```python
"""Driver Code"""
if __name__ == "__main__":
    nums = [4, 1, 3, 1, 5, 2]
    selection_sort(nums)
    print("选择排序完成后 nums =", nums)
```
