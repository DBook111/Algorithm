```python
def solve_hanota(A: list[int], B: list[int], C: list[int]):
    """"""
    n = len(A)
    #  A  n  B  C
    dfs(n, A, B, C)
    
def dfs(i: int, src: list[int], buf: list[int], tar: list[int]):
    """f(i)"""
    #  src  ,  tar
    if i == 1:
        move(src, tar)
        return
    #  f(i-1)  src  i-1  tar  buf
    dfs(i - 1, src, tar, buf)
    #  f(1)  src  tar
    move(src, tar)
    #  f(i-1)  buf  i-1  src  tar
    dfs(i - 1, buf, src, tar)        
        
def move(src: list[int], tar: list[int]):
    """"""
    pan = src.pop()
    tar.append(pan)        
```
