```python
import random

def random_access(nums: list[int]) -> int:
    """"""
    #  [0, len(nums)-1] 
    random_index = random.randint(0, len(nums) - 1)
    # 
    random_num = nums[random_index]
    return random_num

"""Driver Code"""
if __name__ == "__main__":
    # 
    nums = [1, 3, 2, 5, 4]
    print(" nums =", nums)

    # 
    random_num: int = random_access(nums)
    print(" nums ", random_num)
```

```python
def insert(nums: list[int], num: int, index: int):
    """ index  num"""
    #  index 
    for i in range(len(nums)-1, index, -1):
        nums[i] = nums[i - 1]
    nums[index] = num

"""Driver Code"""
if __name__ == "__main__":
    # 
    nums = [1, 3, 2, 5, 4]
    print(" nums =", nums)

    # 
    insert(nums, 6, 3)
    print(" 3  6  nums =", nums)  
```

```python
def remove(nums: list[int], index: int):
    for i in range(index, len(nums)-1):
        nums[i] = nums[i+1]
    nums.pop(len(nums)-1)

"""Driver Code"""
if __name__ == "__main__":
    # 
    nums = [1, 3, 2, 5, 4]
    print(" nums =", nums)

    # 
    remove(nums, 2)
    print(" 2  nums =", nums)
```

```python
def traverse(nums: list[int]):
    """"""
    count = 0
    for i in range(len(nums)):
        count += nums[i]
    for num in nums:
        count += num
    for i, num in enumerate(nums):
        count += nums[i]
        count += num
```

```python
def find(nums: list[int], target: int) -> int:
    """"""
    for i in range(len(nums)):
        if nums[i] == target:
            return i
    return -1

"""Driver Code"""
if __name__ == "__main__":
    # 
    nums = [1, 3, 2, 5, 4]
    print(" nums =", nums)

    # 
    index: int = find(nums, 3)
    print(" nums  3  =", index)
```

```python
def extend(nums: list[int], enlarge: int) -> list[int]:
    """"""
    # 
    res = [0] * (len(nums) + enlarge)
    # 
    for i in range(len(nums)):
        res[i] = nums[i]
    # 
    return res

"""Driver Code"""
if __name__ == "__main__":
    # 
    nums = [1, 3, 2, 5, 4]
    print(" nums =", nums)

    # 
    nums = extend(nums, 3)
    print(" 8  nums =", nums)
```
