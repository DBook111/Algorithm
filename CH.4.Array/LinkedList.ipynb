```python
class ListNode:
    """"""
    def __init__(self, val: int):
        self.val = val
        self.next: ListNode | None = None
```

```python
#  1 -> 3 -> 2 -> 5 -> 4
# 
n0 = ListNode(1)
n1 = ListNode(3)
n2 = ListNode(2)
n3 = ListNode(5)
n4 = ListNode(4)
# 
n0.next = n1
n1.next = n2
n2.next = n3
n3.next = n4
```

```python
def insert(n0: ListNode, P: ListNode):
    """ n0  P"""
    n1 = n0.next
    P.next = n1
    n0.next = P

if __name__ == "__main__":
    # 
    # 
    n0 = ListNode(1)
    n1 = ListNode(3)
    n2 = ListNode(2)
    n3 = ListNode(5)
    n4 = ListNode(4)
    # 
    n0.next = n1
    n1.next = n2
    n2.next = n3
    n3.next = n4
    
    # 
    p = ListNode(0)
    insert(n0, p)
```

```python
def remove(n0: ListNode):
    """ n0 """
    if not n0.next:
        return
    
    P = n0.next
    n0.next = P.next
```

```python
def access(head: ListNode, index: int) -> ListNode | None:
    """ index """
    for i in range(index):
        if not head:
            return None
        head = head.next
    return head

"""Driver Code"""
if __name__ == "__main__":
    # 
    # 
    n0 = ListNode(1)
    n1 = ListNode(3)
    n2 = ListNode(2)
    n3 = ListNode(5)
    n4 = ListNode(4)
    # 
    n0.next = n1
    n1.next = n2
    n2.next = n3
    n3.next = n4

    # 
    node = access(n0, 3)
    print(" 3  = {}".format(node.val))
```

```python
def find(head: ListNode, target: int) -> int:
    """ target """
    index = 0
    while head:
        if head.val == target:
            return index
        head = head.next
        index += 1
    return -1

"""Driver Code"""
if __name__ == "__main__":
    # 
    # 
    n0 = ListNode(1)
    n1 = ListNode(3)
    n2 = ListNode(2)
    n3 = ListNode(5)
    n4 = ListNode(4)
    # 
    n0.next = n1
    n1.next = n2
    n2.next = n3
    n3.next = n4

    # 
    index = find(n0, 2)
    print(" 2  = {}".format(index))
        
```
