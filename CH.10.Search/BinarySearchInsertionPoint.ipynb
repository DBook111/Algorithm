```python
def binary_search_insertion_simple(nums: list[int], target: int) -> int:
    length = len(nums) - 1
    for i in range(length, -1, -1):
        if nums[i] >= target:
            i -= 1
            continue
        return i + 1
```

```python
"""Driver Code"""
if __name__ == "__main__":
    # 无重复元素的数组
    nums = [1, 3, 6, 8, 12, 15, 23, 26, 31, 35]
    # 二分查找插入点
    target = 6
    index = binary_search_insertion_simple(nums, target)
    print(f"元素 {target} 的插入点的索引为 {index}")
```

```python
def binary_search_insertion(nums: list[int], target: int) -> int:
    """二分查找插入点（存在重复元素）"""
    i, j = 0, len(nums) - 1
    while i <= j:
        m = (i + j) // 2
        if nums[m] < target:
            i = m + 1
        elif nums[m] > target:
            j = m - 1
        else:
            j = m - 1
    return i
```

```python
"""Driver Code"""
if __name__ == "__main__":
    # 包含重复元素的数组
    nums = [1, 3, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 10, 12, 15]
    # 二分查找插入点
    target = 6
    index = binary_search_insertion(nums, target)
    print(f"元素 {target} 的插入点的索引为 {index}")
```
