```python
def binary_search(nums: list[int], target: int) -> int:
    """"""
    #  [0, n-1]  i, j 
    i, j = 0, len(nums) - 1
    # 
    while i <= j:
        me = (i + j) // 2
        if nums[me] > target:
            j = me - 1
        elif nums[me] < target:
            i = me + 1
        else:
            return me
    return -1
```

```python
"""Driver Code"""
if __name__ == "__main__":
    target = 6
    nums = [1, 3, 6, 8, 12, 15, 23, 26, 31, 35]

    # 
    index = binary_search(nums, target)
    print(" 6  = ", index)
```

```python
def binary_search_lcro(nums: list[int], target: int) -> int:
    """"""
    #  [0, n)  i, j +1
    i, j = 0, len(nums)
    #  i = j 
    while i < j:
        m = (i + j) // 2  #  m
        if nums[m] < target:
            i = m + 1  #  target  [m+1, j) 
        elif nums[m] > target:
            j = m  #  target  [i, m) 
        else:
            return m  # 
    return -1  #  -1
```
