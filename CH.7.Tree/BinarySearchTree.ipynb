```python
class TreeNode:
    def __init__(self, val: int) -> None:
        self.val = val
        self.left: TreeNode | None = None
        self.right: TreeNode | None = None

def search(self, num: int) -> TreeNode | None:
    cur = self._root
    while cur is not None:
        if cur.val < num:
            cur = cur.right
        elif cur.val > num:
            cur = cur.left
        else:
            break
    return cur
```

```python
def insert(self, num: int):
    """"""
    # 
    if self._root is None:
        self._root = TreeNode(num)
        return
    # 
    cur, pre = self._root, None
    while cur is not None:
        # 
        if cur.val == num:
            return
        pre = cur
        if cur.val < num:
            cur = cur.right
        else:
            cur = cur.left
    node = TreeNode(num)
    if pre.val < num:
        pre.right = node
    else:
        pre.left = node
```
