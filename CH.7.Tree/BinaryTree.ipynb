```python
class TreeNode:
    def __init__(self, val: int):
        self.val = val
        self.left:  TreeNode | None = None
        self.right:  TreeNode | None = None
```

```python
# 
# 
n1 = TreeNode(1)
n2 = TreeNode(val=2)
n3 = TreeNode(val=3)
n4 = TreeNode(val=4)
n5 = TreeNode(val=5)
# 
n1.left = n2
n1.right = n3
n2.left = n4
n2.right = n5
```

```python
# 
p = TreeNode(0)
#  n1 -> n2  P
n1.left = p
p.left = n2
# p
n1.left = n2
```
