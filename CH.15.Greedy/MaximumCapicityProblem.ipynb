```python
def max_capacity(ht: list[int]) -> int:
    """"""
    #  i, j
    i, j = 0, len(ht) - 1
    #  0
    res = 0
    # 
    while i < j:
        # 
        cap = min(ht[i], ht[j]) * (j - i)
        res = max(res, cap)
        # 
        if ht[i] < ht[j]:
            i += 1
        else:
            j -= 1
    return res
```

```python
"""Driver Code"""
if __name__ == "__main__":
    ht = [3, 8, 5, 2, 7, 7, 3, 4]

    # 
    res = max_capacity(ht)
    print(f" {res}")
```
