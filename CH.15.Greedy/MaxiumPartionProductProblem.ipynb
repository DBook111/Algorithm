```python
import math

def max_product_cutting(n: int) -> int:
    """"""
    #  n <= 3  1
    if n <= 3:
        return 1 * (n - 1)
    #  3 a  3 b 
    a, b = n // 3, n % 3
    if b == 1:
        #  1  1 * 3  2 * 2
        return int(math.pow(3, a - 1)) * 2 * 2
    if b == 2:
        #  2 
        return int(math.pow(3, a)) * 2
    #  0 
    return int(math.pow(3, a))
```
